```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def create_2d_state(state: list, colors: list, n=4):
    state_2d = np.zeros((n, n), dtype=int)
    color_map = {pos: color for pos, color in zip(state, colors)}
    for x, y in state:
        state_2d[x-1, y-1] = color_map.get((x, y), 0)
    return state_2d

def plot_2d(state, colors, title, n=4):
    state_2d = create_2d_state(state, colors, n)
    fig, ax = plt.subplots()
    cmap = mcolors.ListedColormap(['white', 'red', 'blue', 'green', 'yellow'])
    bounds = [0, 1, 2, 3, 4, 5]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)
    ax.imshow(state_2d, cmap=cmap, norm=norm, origin='upper')
    ax.set_title(title)
    ax.set_xticks(np.arange(n))
    ax.set_yticks(np.arange(n))
    ax.set_xticklabels(np.arange(1, n+1))
    ax.set_yticklabels(np.arange(1, n+1))
    ax.set_xticks(np.arange(n+1)-0.5, minor=True)
    ax.set_yticks(np.arange(n+1)-0.5, minor=True)
    ax.grid(which='minor', color='black', linestyle='-', linewidth=2)
    plt.show()

# Example usage
state = [(1, 1), (2, 2), (3, 3), (4, 4)]  # (row, col)
colors = [1, 1, 2, 2]  # Color indices for each state
plot_2d(state, colors, "2D Grid Plot")
```

```python
# Example usage
state = [(1, 1), (1, 2), (1, 3), (1, 4)]  # (row, col)
colors = 5 * np.ones(4)  # Color indices for each state
plot_2d(state, colors, "2D Grid Plot")
```

```python

number_box = 16
state_1d = []

def create_1d_state(state: list, colors: list, n=number_box):
    state_1d = np.zeros(n, dtype=int)
    color_map = {pos: color for pos, color in zip(state, colors)}
    for x in state:
        state_1d[x-1] = color_map.get(x, 0)
    return state_1d

def plot_1d(state, colors, title, n=number_box):  #by cha
    state_1d = create_1d_state(state, colors, n)
    fig, ax = plt.subplots(figsize=(n, 1))
    cmap = mcolors.ListedColormap(['white', 'red', 'blue', 'green', 'yellow', 'grey'])
    bounds = [0, 1, 2, 3, 4, 5]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)
    ax.imshow(state_1d.reshape(1, -1), cmap=cmap, norm=norm, aspect='auto')
    ax.set_title(title)
    ax.set_xticks(np.arange(n))
    ax.set_xticklabels(np.arange(1, n+1))
    ax.set_yticks([])
    ax.set_xticks(np.arange(n+1)-0.5, minor=True)
    ax.grid(which='minor', color='black', linestyle='-', linewidth=2)
    plt.show()

for i in range (number_box):
    state_1d.append(i+1)  # Position in 1D
colors_1d = 5* np.ones(number_box) # Color indices for each position
plot_1d(state_1d, colors_1d, "1D Grid Plot")
```
